Python zipfile Module 


The zipfile module is a built-in Python module used to create, read, write, extract, and manage ZIP archives.

A ZIP archive is a compressed file that stores one or more files together, reducing storage space and making files easier to share.

Definition:

The zipfile module provides tools to create, read, write, extract, and inspect ZIP (.zip) archives in Python.

2. Why Do We Use zipfile?

Without zipfile:

Compressing files manually is difficult.
Extracting files requires external software.

With zipfile:
    Compress files.
    Extract files.
Read files directly without extracting.
    Create backups.
    Share multiple files as one ZIP.

3. Real-Life Applications

    File backup systems

    Software installation packages

    Email attachments

    Android APK files (APK is a ZIP archive)

    Office files (.docx, .xlsx, .pptx) are ZIP archives internally

    Game assets
    
    Cloud storage

5. Opening a ZIP File

     zipfile.ZipFile(file, mode)

| Mode  | Meaning                              |
| ----- | ------------------------------------ |
| `"r"` | Read existing ZIP                    |
| `"w"` | Create new ZIP (overwrites existing) |
| `"a"` | Append files                         |
| `"x"` | Create new ZIP (error if exists)     |


```python
# example

import zipfile

zip_file = zipfile.ZipFile("Day5.zip", "r")

zip_file.close()
```

6. Compression Types

| Constant       | Compression                      |
| -------------- | -------------------------------- |
| `ZIP_STORED`   | No compression                   |
| `ZIP_DEFLATED` | Standard compression (most used) |
| `ZIP_BZIP2`    | Better compression               |
| `ZIP_LZMA`     | Highest compression              |

Example:-

```python
zipfile.ZipFile(
    "data.zip",
    "w",
    compression=zipfile.ZIP_DEFLATED
)
```

All Important Methods

1. ZipFile()

Purpose

Open or create a ZIP archive.

```python
import zipfile

z = zipfile.ZipFile("Day5.zip", "w")
z.close()
```

2. write()

Purpose

Adds an existing file to the ZIP archive.

```python
import zipfile

with zipfile.ZipFile("Day5.zip", "w") as z:
    z.write("hello.txt")
```

Output

demo.zip

 └── hello.txt

3. writestr()

Purpose

Creates a file inside the ZIP from a string.

```python
import zipfile

with zipfile.ZipFile("demo.zip", "w") as z:
    z.writestr("message.txt",
               "Hello Everyone!")
```

No physical file is required.

4. read()

Purpose

Reads the contents of a file inside the ZIP.

```python
with zipfile.ZipFile("demo.zip","r") as z:
    data = z.read("message.txt")

print(data)
```

Output

b'Hello Everyone!'

```python
#  Convert bytes to text 

print(data.decode())

#  Output 

#  Hello Everyone! 
```

5. open()

Purpose

Opens a file inside the ZIP without extracting it.

```python
with zipfile.ZipFile("demo.zip") as z:

    with z.open("message.txt") as f:

        print(f.read().decode())
```

Difference

read() → Reads everything immediately.

open() → Returns a file object.

Useful for large files.

6. extract()

Purpose

Extracts one file.

```python
with zipfile.ZipFile("demo.zip") as z:

    z.extract("message.txt")
```

7. extractall()

Purpose

Extracts all files.

```python
with zipfile.ZipFile("demo.zip") as z:

    z.extractall("output")
```

8. namelist()

Purpose

Returns a list of filenames.

```python
with zipfile.ZipFile("demo.zip") as z:

    print(z.namelist())
```

9. printdir()

Purpose

Prints file details.

```python
with zipfile.ZipFile("demo.zip") as z:

    z.printdir()
    
# File Name      Modified Size
# hello.txt
# message.txt
```

10. infolist()

Purpose

Returns information about every file.

```python
with zipfile.ZipFile("demo.zip") as z:

    info = z.infolist()

    for file in info:
        print(file.filename)
        print(file.file_size)
```

hello.txt
25

message.txt
15

11. getinfo()

Purpose

Returns information for one file.

```python
with zipfile.ZipFile("demo.zip") as z:

    info = z.getinfo("hello.txt")

print(info.filename)
print(info.file_size)
```

12. testzip()

Purpose

Checks archive integrity.

```python
with zipfile.ZipFile("demo.zip") as z:

    result = z.testzip()

print(result)
```

Output

None

None means there are no corrupted files.

If a file is corrupted, it returns that filename.

13. setpassword()

Purpose

Sets a password for reading encrypted ZIP files.

```python
with zipfile.ZipFile("secure.zip") as z:

    z.setpassword(b"1234")

    print(z.read("data.txt"))
```

Important: Python's zipfile can read encrypted ZIPs but cannot create password-protected ZIP files directly.

14. close()

Purpose

Closes the ZIP archive.

```python
z = zipfile.ZipFile("demo.zip")

z.close()
```

Better approach

```python
with zipfile.ZipFile("demo.zip") as z:

    print(z.namelist())
```

The with statement closes the file automatically.
