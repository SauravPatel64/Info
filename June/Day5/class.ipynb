Methods of named Tuple
1) _fields :- it return field name
2) _asdict()  :- convert nameddtuple into dictionary
3) _replace() :- It creates a new object with updated value
4) _make() :- it is a class method available in named tuple it create namedtuple stance from an iterable

```python
# _fields

from collections import namedtuple
Student = namedtuple("Student",["rollno","name","marks"])
s1 = Student("121","Saurav","98")
print(s1)
print(Student._fields)
```

```python
# _asdict
from collections import namedtuple
Student = namedtuple("Student",["rollno","name","marks"])
s1 = Student("121","Saurav","98")
print(s1)
print(s1._asdict())
```

```python
# _makes()

from collections import namedtuple
Student = namedtuple("Student",["rollno","name","marks"])
data = ["121","Saurav","98"]
s1 =  Student._make(data)
print(s1)
print(s1.rollno)
print(s1.name)
print(s1.marks)
```
